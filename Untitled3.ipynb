```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras import Input
from keras.datasets import mnist
from keras.layers import Conv2D,Dense,MaxPooling2D,UpSampling2D,Conv2DTranspose
from keras.models import Model
from keras.preprocessing import image

# Load and preprocess the MNIST dataset
(X_train,_),(X_test,_)=mnist.load_data()

def pre_process(X):
  """Preprocess the input data."""
  X=X/255
  X=X.reshape(len(X),784)
  return X

X_train=pre_process(X_train)
X_test=pre_process(X_test)

# Define model parameters
input_dim,output_dim=784,784
encod_dim=100
hidden_dim=256

# Define the autoencoder architecture
input_layer=Input(shape=(input_dim,),name="Input")
hidden_layer_1=Dense(hidden_dim,activation="relu",name='Hidden_1')(input_layer)
code_layer=Dense(encod_dim,activation="relu",name="code")(hidden_layer_1)
hidden_layer_2=Dense(hidden_dim,activation='relu',name="Decode")(code_layer)
output_layer=Dense(output_dim,activation="sigmoid",name="Output")(hidden_layer_2)

# Compile and summarize the autoencoder model
AE=Model(input_layer,output_layer)
AE.compile(optimizer="adam",loss="binary_crossentropy")
AE.summary()

# Train the autoencoder model
AE.fit(X_train,X_train,epochs=10)
```

```python
# Create a model to extract encoded data
get_encoded_data = Model(inputs=AE.input, outputs=AE.get_layer("code").output)
```

```python
# Obtain decoded and encoded data
decoded_data = AE.predict(X_test)
encoded_data = get_encoded_data.predict(X_test)
```

```python
def show_data(X, n=10, height=28, width=28, title=""):
    plt.figure(figsize=(10, 3))
    for i in range(n):
        ax = plt.subplot(2,n,i+1)
        plt.imshow(X[i].reshape((height,width)))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.suptitle(title, fontsize = 20)
show_data(X_test, title="original data")
show_data(encoded_data, height=10, width=10, title="encoded data")
show_data(decoded_data, title="decoded data")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras import Input
from keras.datasets import mnist
from keras.layers import Conv2D, Dense, MaxPooling2D, UpSampling2D, Conv2DTranspose
from keras.models import Model
from keras.preprocessing import image

# Load the MNIST dataset
(X_train, _), (X_test, _) = mnist.load_data()

# Define the input layer
input_layer = Input(shape=(28, 28, 1), name="input")

# Define the encoder layers
x = Conv2D(8, (3, 3), activation='relu', padding='same')(input_layer)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(4, (3, 3), activation='relu', padding='same')(x)
code_layer = MaxPooling2D((2, 2), name="Code")(x)

# Define the decoder layers
x = Conv2DTranspose(4, (3, 3), activation='relu', padding="same")(code_layer)
x = UpSampling2D((2, 2))(x)
x = Conv2DTranspose(8, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
output_layer = Conv2D(1, (3, 3), padding='same', name='output')(x)

# Create the autoencoder model
cat_AE = Model(input_layer, output_layer)
cat_AE.compile(optimizer='adam', loss='mse')
cat_AE.summary()

# Train the autoencoder model
cat_AE.fit(X_train, X_train,
           epochs=30,
           batch_size=32,
           shuffle=True,
           validation_data=(X_test, X_test))
```

```python
coded_model = Model(cat_AE.input, cat_AE.get_layer("Code").output)
coded_data = coded_model.predict(X_test)
decoded_img = cat_AE.predict(X_test)

n = 10  # number of images to display
plt.figure(figsize=(20, 6))
for i in range(n):
    # Original images
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(X_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Encoded images
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(decoded_img[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Decoded images
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(coded_data[i].reshape(7, 7 * 4).T)  # Corrected reshaping
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

plt.show()

```
